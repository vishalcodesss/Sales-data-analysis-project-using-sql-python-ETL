```python
import zipfile                  # used to unzip the file downloaded from kaggle
zip_ref = zipfile.ZipFile('orders.csv.zip') 
zip_ref.extractall()
zip_ref.close()
```

```python
import pandas as pd
# read data from the files and handle null values
df= pd.read_csv('orders.csv',na_values = ['Not Available','unknown'])
```

```python
df.head(20)
df['Ship Mode'].unique()
```

```python
# rename columns names and make them lowercase and replace spaces with underscore
# df.rename(columns = {'Order Id':'order_id','Ship Mode':'ship_mode'})   # this process is very lengthy so we try to avoid it
df.columns = df.columns.str.lower()        # converted into lower case
df.columns = df.columns.str.replace(' ', '_')   # replaced spaces with underscore
df.head()
```

```python
         # derive new columns - discount, sale price and profit
df['discount'] = df['discount_percent']/100 * df['list_price']
df['sale_price'] = df['list_price'] - df['discount']
df['Profit'] = df['sale_price'] - df['cost_price']
df.head()
```

```python
                # convert order date from object data type to datetime
df['order_date'] = pd.to_datetime (df['order_date'], format= "%Y-%m-%d")
df
```

```python
                        # drop cost price, list price and discount percent columns
df.drop(columns=['cost_price','list_price','discount_percent'], inplace = True)
df
```
